```python
def get_sections(path):
    response = requests.get(path)
    soup = BeautifulSoup(response.text, 'html.parser')
    noshow_spans = soup.find_all('span', class_='noshow')
    if path == "https://www.perseus.tufts.edu/hopper/text?doc=Perseus%3atext%3a1999.01.0169%3atext%3dEuthyph.":
        section_texts = [span.get_text()[-3:].strip() for span in noshow_spans if span.get_text().startswith("sec")]
    else:
        section_texts = [span.get_text()[-4:].strip() for span in noshow_spans if span.get_text().startswith("sec")]
    return section_texts
    
```

```python
plato_information = []
for i in range(len(plato_links)):
    if names[i] in "Alc.+2":
        link = "https://www.perseus.tufts.edu/hopper/text?doc=Perseus%3Atext%3A1999.01.0175%3Atext%3DAlc.+2%3Asection%3D138a"
        plato_information.append([names[i], plato_links[i], get_sections(link)])
    elif names[i] == "Hipparch.": 
        link= "https://www.perseus.tufts.edu/hopper/text?doc=Perseus%3Atext%3A1999.01.0175%3Atext%3DHipparch.%3Asection%3D225a"
        plato_information.append([names[i], plato_links[i], get_sections(link)])
    else:
        plato_information.append([names[i], plato_links[i], get_sections(plato_links[i])])
print(plato_information)
```

Nachdem wir nun f√ºr jedes Werk seine Sektionen haben, k√∂nnnen wir den jeweiligen Text der Sektion laden und speichern. 

```python
def get_text(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    greek_texts = soup.find_all('p')
    greek_texts = [p.get_text() for p in greek_texts]
    return " ".join(greek_texts)
```

```python
def save_as_dataframe(data, name):
    data = pd.DataFrame(data, columns=['section', 'path', 'text'])
    data.to_csv(name+'.csv')
    
```
